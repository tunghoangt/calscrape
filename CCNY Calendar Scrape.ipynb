```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
from IPython.display import HTML
```

```python
req = requests.get('https://www.ccny.cuny.edu/registrar/fall-2018-academic-calendar')
```

```python
req.text
```

```python
soup = BS(req.text,'html.parser')
```

```python
print(soup.prettify)
```

```python
dir(soup)
```

```python
soup.title
```

```python
table = soup.find_all('table')
```

```python
table
```

```python
HTML(str(table))
```

```python
table[0].find_all('tr')
```

```python
rows = [row for row in table[0].find_all('tr')]
```

```python
colheads = [col.get_text() for col in rows[0].find_all('p')][0:2]
```

```python
colheads.append('DESCRIPTION')
colheads
```

```python
value_rows = [ [ col.get_text().strip().replace('\n',' ').replace('\xa0','') for col in row.find_all('td') ] for row in rows[1:]]
value_rows = [value for value in value_rows if value[1] != '']
value_rows
```

```python
value_series =list(zip(*value_rows))
value_series
```

```python
data_dict = dict(zip(colheads,value_series))
data_dict
```

```python
df = pd.DataFrame(data_dict)
```
